```python
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
import pandas as pd
import datetime
from datetime import datetime
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
from collections import Counter
plt.rcParams.update({'font.size': 22})
import sys
```

```python
swarms = pd.read_csv('swarms_per_day_final', delim_whitespace=True, names=['Date', 'No.'])
swarms.head()

d = list(swarms['Date'])
#d = [datetime.strptime(dte, "%Y/%m/%d") for dte in d]
n = list(swarms['No.'])
n = [int(no) for no in n]

```

```python
fig, ax = plt.subplots(figsize=(12,8))
plt.plot(d, n, color='palevioletred')
plt.xlabel('Fecha (AAAA/MM/DD)')
plt.ylabel('Número de sismos por día')
plt.xticks(rotation=80)
plt.rcParams['xtick.labelsize'] = 18
ax.xaxis.set_major_locator(plt.MaxNLocator(14))
plt.title('Posibles enjambres sísmicos por día (Nov 2017- Nov 2018)')
```
